# **Competition Code**

## 1. Preprocessing Steps

```python
import os
import json
import random
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

import torch
from torch.utils.data import Dataset, DataLoader

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    get_linear_schedule_with_warmup,
)

from torch.optim import AdamW  

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)
```

##### Read Data

```python
DATA_DIR = "./competition"  

with open(os.path.join(DATA_DIR, "final_posts.json"), "r", encoding="utf-8") as f:
    posts_raw = json.load(f)

#  DataFrame：id / text
post_list = []
for p in posts_raw:
    post = p["root"]["_source"]["post"]
    post_list.append({"id": post["post_id"], "text": post["text"]})
posts_df = pd.DataFrame(post_list)

id_df = pd.read_csv(os.path.join(DATA_DIR, "data_identification.csv"))  # id, split(train/test)
emo_df = pd.read_csv(os.path.join(DATA_DIR, "emotion.csv"))             # train label
sample_sub = pd.read_csv(os.path.join(DATA_DIR, "samplesubmission.csv"))

# combine text, split, and emotion
data = id_df.merge(posts_df, on="id", how="left")
data = data.merge(emo_df, on="id", how="left")  # test emotion: NaN

train_df = data[data["split"] == "train"].reset_index(drop=True)
test_df = data[data["split"] == "test"].reset_index(drop=True)

print("Train size:", len(train_df))
print("Test  size:", len(test_df))
print("\nLabel Distribution：")
print(train_df["emotion"].value_counts())
```

##### Data Exploration

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# Label distribution
plt.figure(figsize=(6, 4))
order = train_df['emotion'].value_counts().index
sns.countplot(data=train_df, x='emotion', order=order)
plt.title("Emotion label distribution (train)")
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Tweet length distribution 
train_df['char_len'] = train_df['text'].str.len()
test_df['char_len']  = test_df['text'].str.len()

plt.figure(figsize=(6, 4))
sns.histplot(train_df['char_len'], bins=40, kde=True)
plt.title("Tweet length distribution (characters) – train")
plt.xlabel("Length (chars)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Tweet length vs emotion
plt.figure(figsize=(7, 4))
sns.boxplot(data=train_df, x='emotion', y='char_len', order=order)
plt.title("Tweet length by emotion (characters)")
plt.xlabel("Emotion")
plt.ylabel("Length (chars)")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# a few example tweets
for emo in train_df['emotion'].unique():
    print(f"\n=== Emotion: {emo} ===")
    examples = train_df[train_df['emotion'] == emo]['text'].dropna().head(3).tolist()
    for i, t in enumerate(examples, 1):
        print(f"{i}. {t}")
```

##### Take Away:
1. Some emotions appear much more often(joy, sadness might dominant).

4. For anger emotion, the lengths of text are slightly longer.

## 2. Feature Engineering Steps

1. turning string labels into numeric IDs.
2. defining a consistent mapping, and passing id2label / label2id into the model.

```python
labels = sorted(train_df["emotion"].unique())

# create mappings
label2id = {lab: i for i, lab in enumerate(labels)}
id2label = {i: lab for lab, i in label2id.items()}

num_labels = len(labels)

# mapping in dataset
train_df["label"] = train_df["emotion"].map(label2id)
```

```python
# train / valid split
train_df, valid_df = train_test_split(
    train_df,
    test_size=0.1,
    random_state=42,
    stratify=train_df["label"],
)

print("Train:", len(train_df), "Valid:", len(valid_df))
```

## 3. Model Implementation Steps

##### BERT

"bert-base-uncased" loads a full pretrained BERT (12 layers, 768 hidden size), turning each token sequence into contextual hidden states.

```python
PRETRAINED_MODEL = "bert-base-uncased"
MAX_LEN = 128
BATCH_SIZE = 16

tokenizer = AutoTokenizer.from_pretrained(PRETRAINED_MODEL)

class EmotionDataset(Dataset):
    def __init__(self, df, tokenizer, max_len, is_test=False):
        self.texts = df["text"].tolist()
        self.is_test = is_test
        if not is_test:
            self.labels = df["label"].tolist()
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        encoded = self.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt",
        )
        item = {k: v.squeeze(0) for k, v in encoded.items()}
        if not self.is_test:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

train_dataset = EmotionDataset(train_df, tokenizer, MAX_LEN, is_test=False)
valid_dataset = EmotionDataset(valid_df, tokenizer, MAX_LEN, is_test=False)
test_dataset  = EmotionDataset(test_df,  tokenizer, MAX_LEN, is_test=True)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)
```

##### Classifier

AutoModelForSequenceClassification automatically attaches a small classifier on top of BERT:

* It takes the [CLS] token representation.

* It applies a linear layer to produce num_labels logits.

* During training, outputs.loss is computed with CrossEntropyLoss between logits and your integer labels.

* AdamW updates all BERT + head parameters with a small learning rate, mild weight decay.

* The scheduler does 10% warmup then linear decay (a standard for transformer fine-tuning).

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

model = AutoModelForSequenceClassification.from_pretrained(
    PRETRAINED_MODEL,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
)
model.to(device)

# optimizer & scheduler
EPOCHS = 3
optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=1e-2)


num_training_steps = EPOCHS * len(train_loader)
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(0.1 * num_training_steps),
    num_training_steps=num_training_steps,
)
```

##### Training and Validation

```python
from sklearn.metrics import accuracy_score

def train_one_epoch(epoch):
    model.train()
    total_loss = 0.0
    y_true, y_pred = [], []

    for batch in train_loader:
        batch = {k: v.to(device) for k, v in batch.items()}

        optimizer.zero_grad()
        outputs = model(**batch)
        loss = outputs.loss
        logits = outputs.logits

        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()

        preds = torch.argmax(logits, dim=-1).detach().cpu().numpy()
        labels = batch["labels"].detach().cpu().numpy()
        y_true.extend(labels)
        y_pred.extend(preds)

    acc = accuracy_score(y_true, y_pred)
    print(f"[Train] Epoch {epoch} | loss = {total_loss/len(train_loader):.4f} | accuracy = {acc:.4f}")

def eval_one_epoch(epoch):
    model.eval()
    total_loss = 0.0
    y_true, y_pred = [], []

    with torch.no_grad():
        for batch in valid_loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            loss = outputs.loss
            logits = outputs.logits

            total_loss += loss.item()

            preds = torch.argmax(logits, dim=-1).detach().cpu().numpy()
            labels = batch["labels"].detach().cpu().numpy()
            y_true.extend(labels)
            y_pred.extend(preds)

    acc = accuracy_score(y_true, y_pred)
    print(f"[Valid] Epoch {epoch} | loss = {total_loss/len(valid_loader):.4f} | accuracy = {acc:.4f}")
    return acc

best_acc = 0.0
for epoch in range(1, EPOCHS + 1):
    train_one_epoch(epoch)
    val_acc = eval_one_epoch(epoch)
    if val_acc > best_acc:
        best_acc = val_acc
        torch.save(model.state_dict(), "best_bert_emotion.pt")
        print(f"Best Model (acc={best_acc:.4f})")

# 載入最佳模型做預測
model.load_state_dict(torch.load("best_bert_emotion.pt", map_location=device))
model.to(device)
model.eval()

# submission
all_test_preds = []

with torch.no_grad():
    for batch in test_loader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=-1).detach().cpu().numpy()
        all_test_preds.extend(preds.tolist())

test_emotions = [id2label[idx] for idx in all_test_preds]

submission = pd.DataFrame({
    "id": test_df["id"].values,
    "emotion": test_emotions,
})

submission.to_csv("submission_bert.csv", index=False)
print("已輸出 submission_bert.csv")
```
